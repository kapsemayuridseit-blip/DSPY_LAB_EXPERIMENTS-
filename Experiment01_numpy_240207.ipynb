**Array Slicing: Accessing Subarrays**
Just as we can use square brackets to access individual array elements, we can also use them to access subarrays with the slice notation, marked by the colon (:) character. The NumPy slicing syntax follows that of the standard Python list; to access a slice of an array x, use this:

x[start:stop:step]
If any of these are unspecified, they default to the values **start=0, stop=size of dimension, step=1**. We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions

**Introducing UFuncs**
For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.

```python
#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```python
#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
```

Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

Make changes in 1 perticular value in the subObjCpy. Check values of Obj subObjCpy to check impact of this change of element in original and copied object. What do you conclude?

In NumPy, slicing creates a VIEW, not a deep copy

So changes in subObjCpy may affect the original array

```python
import numpy as np

# Create original 3×4×2 array
Obj = np.random.randint(1, 51, size=(3, 4, 2))

# Reshape to 4×6
New_Obj = Obj.reshape(4, 6)

print("Original 4×6 Array:")
print(New_Obj)

# Take 2×3 sub-array from left bottom
subObjCpy = New_Obj[-2:, :3]

print("\nSub-array (Before Change):")
print(subObjCpy)

# Change one particular value in subObjCpy
subObjCpy[0, 0] = 999

print("\nSub-array (After Change):")
print(subObjCpy)

print("\nOriginal 4×6 Array (After Change):")
print(New_Obj)

print("\nOriginal 3×4×2 Array (After Change):")
print(Obj)
```

The value 999 appears:

in subObjCpy

in New_Obj

and also in original Obj

subObjCpy is not an independent copy.
It is a view of the original array
Therefore, changing an element in subObjCpy also changes the original array (Obj).

**Q7**If you were to filter sub array view from above Obj object and want to apply modification in original and/or sub obj change respective element in other how will you achieve it?

**Ans**Filter a sub-array from original NumPy array Obj and control whether changes in one reflect in the other or not.

```python
import numpy as np

# Step 1: Create original 3×4×2 array
Obj = np.random.randint(1, 51, size=(3, 4, 2))

# Step 2: Reshape to 4×6
reshaped = Obj.reshape(4, 6)

print("Original 4×6 Array:\n", reshaped)

# ---------------- VIEW (Slicing) ----------------
# Creates a VIEW → changes propagate both ways
sub_view = reshaped[-2:, :3]
sub_view[0, 0] = 999

print("\nAfter modifying VIEW:")
print("Sub View:\n", sub_view)
print("Original Array:\n", reshaped)

# ---------------- COPY ----------------
# Independent copy → no propagation
sub_copy = reshaped[-2:, :3].copy()
sub_copy[0, 1] = 888

print("\nAfter modifying COPY:")
print("Sub Copy:\n", sub_copy)
print("Original Array:\n", reshaped)

# ---------------- FILTER (Boolean Mask) ----------------
# Filtering creates COPY
mask = reshaped > 30
filtered = reshaped[mask]
filtered[0] = 777   # No effect on original

print("\nAfter modifying FILTERED array:")
print("Filtered Array:\n", filtered)
print("Original Array:\n", reshaped)

# ---------------- FILTER + MODIFY ORIGINAL ----------------
# Apply change to original using same mask
reshaped[mask] = 555

print("\nAfter modifying ORIGINAL using mask:")
print("Original Array:\n", reshaped)
print("Original 3×4×2 Obj:\n", Obj)
```

**Slicing creates a VIEW, so modifications in sub-array reflect in the original array.**

**Boolean filtering and fancy indexing create a COPY, so changes do not propagate.**

**To modify the original array after filtering, apply changes using the same mask or indices.**

**Use .copy() when an independent sub-array is required.**
