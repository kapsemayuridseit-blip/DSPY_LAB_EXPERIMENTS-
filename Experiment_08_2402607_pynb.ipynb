**Q1. What is a Two-Tailed and One-Tailed T-Test?**

Answer:
A T-test is used to determine whether there is a significant difference between the means of groups.

One-tailed T-test:
Used when the hypothesis checks the effect in only one direction (greater than OR less than).
Example: Testing if a new teaching method increases student scores.

Two-tailed T-test:
Used when the hypothesis checks for any difference in either direction (greater than OR less than).
Example: Testing if a new teaching method changes student scores (increase or decrease).

```python
from scipy import stats

# Sample data
sample1 = [70, 75, 80, 85, 90]
sample2 = [68, 74, 78, 82, 88]

# Two-tailed t-test
t_stat, p_value = stats.ttest_ind(sample1, sample2)

print("T statistic:", t_stat)
print("Two-tailed p-value:", p_value)

# One-tailed p-value
one_tailed_p = p_value / 2
print("One-tailed p-value:", one_tailed_p)
```

**Q2.Explain what is Type I and Type II errors with an example.**

Answer:Type I and Type II Errors

In hypothesis testing, we make decisions based on sample data, which can sometimes be wrong.

Type I Error (α – False Positive):
This happens when we reject the null hypothesis even though it is true.
In simple terms, we detect something that does not actually exist.

Type II Error (β – False Negative):
This happens when we fail to reject the null hypothesis even though it is false.
That means we miss a real effect.

Example:
Suppose we test a new drug.

Type I → We say the drug works, but it actually doesn’t.

Type II → We say the drug doesn’t work, but it actually does.

**Q3.What is power of statistical test?**

Answer:Power of a Statistical Test

The power of a statistical test tells us how good the test is at detecting a real effect.

Power=1−β
Power=1−β

It is the probability of correctly rejecting a false null hypothesis.

Higher power means a lower chance of Type II error.

A power of 0.8 (80%) means the test will correctly detect the effect 8 out of 10 times.

Factors affecting power:

Larger sample size → higher power

Larger effect size → higher power

Higher significance level (α) → higher power

**Q4. Explain input and out put characteristics for  stats.ttest_rel function in your own words**

Answer: Input and Output of stats.ttest_rel

The function stats.ttest_rel is used for a paired t-test, which compares two related samples.

Inputs:

a: First set of values (e.g., before treatment)

b: Second set of values (e.g., after treatment)

Both must have the same length because values are paired.

Outputs:

t_statistic: Measures how much the means differ relative to variation

p_value: Probability that the observed difference is due to chance

Decision rule:

If p < 0.05, reject null hypothesis → significant difference

If p ≥ 0.05, no significant difference

```python
from scipy import stats

# Paired data (before and after)
before = [50, 55, 60, 62, 58]
after = [55, 60, 65, 68, 63]

t_stat, p_value = stats.ttest_rel(before, after)

print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Significant difference")
else:
    print("No significant difference")
```

**Q5.What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

Answer:Function for Power of a Statistical Test

In Python, we use TTestPower from statsmodels.stats.power to calculate power.

Function:
TTestPower().power(effect_size, nobs, alpha)

Parameters:

effect_size: Strength of difference between groups

nobs: Number of observations (sample size)

alpha: Significance level (commonly 0.05)

Output:

A value between 0 and 1

Closer to 1 → strong test

Less than 0.5 → weak test

Significance:

Helps decide if sample size is enough

Ensures reliable results

Common goal: power ≥ 0.8

```python
from statsmodels.stats.power import TTestPower

analysis = TTestPower()

power = analysis.power(effect_size=0.5, nobs=30, alpha=0.05)

print("Power of test:", power)

if power >= 0.8:
    print("Good statistical power")
else:
    print("Low statistical power")
```
