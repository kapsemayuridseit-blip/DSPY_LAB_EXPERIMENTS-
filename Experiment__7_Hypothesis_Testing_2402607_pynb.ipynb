**Q1.What is need for Hypothesis Testing?**
Answer:Need for Hypothesis Testing

Hypothesis Testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. In real life, it is often difficult or impossible to collect data from the entire population, so we use a sample and test assumptions about the population.

1.To Make Decisions Using Sample Data
Hypothesis testing helps researchers determine whether the results obtained from a sample are significant or happened by chance.
It provides a scientific way to decide whether a statement about a population is true or false.

2.To Test Assumptions or Claims
Many studies begin with a claim or assumption called a hypothesis.
Hypothesis testing helps verify whether there is enough evidence in the data to support or reject the claim.

Example:Suppose we want to test the claim that the average age of students in TE IT class is 19 years. By collecting sample data and applying hypothesis testing, we can determine whether this claim is valid.

3.To Reduce Uncertainty in Decision Making
When analyzing data, there is always uncertainty. Hypothesis testing uses statistical methods to measure the reliability of results and reduce the chance of making incorrect decisions.

4.To Compare Groups or Methods
It is commonly used to check whether there is a significant difference between two groups or processes.

Example:Comparing two teaching methods to see which improves student performance.
Checking if a new medicine works better than the existing one.

5.To Support Research and Data Analysis
Hypothesis testing is widely used in data science, statistics, machine learning, business analysis, and scientific research to validate results and make data-driven conclusions.

**Q6.Check if you can accept Ho at alpha 95% ? Match your conclusions with python functions.**

Answer:### Checking Whether (H_0) Can Be Accepted at 95% Significance Level

We want to test whether the **average age of TE IT students is 19 years**.

---

### 1. Hypotheses

* **Null Hypothesis (H₀):**
  ( \mu = 19 )

* **Alternative Hypothesis (H₁):**
  ( \mu \ne 19 )

This is a **two-tailed test**.

---

### 2. Level of Significance

Confidence level = **95%**

[
\alpha = 1 - 0.95 = 0.05
]

---

### 3. Sample Data (Example)

Suppose we collected age data of **10 students**:

18, 19, 20, 19, 18, 20, 19, 19, 18, 20

* Sample size (n = 10)
* Sample mean ( \bar{x} = 19 )
* Sample standard deviation ( s \approx 0.82 )

---

### 4. Test Statistic (t-test)

t = \frac{\bar{x}-\mu}{s/\sqrt{n}}

Substitute values:

[
t = \frac{19-19}{0.82/\sqrt{10}}
]

[
t = 0
]

---

### 5. Critical Value

Degrees of freedom:

[
df = n - 1 = 9
]

At **α = 0.05 (two-tailed test)**

[
t_{critical} \approx 2.262
]

Decision rule:

* If ( |t| > 2.262 ) → Reject (H_0)
* If ( |t| \le 2.262 ) → Accept (H_0)

Since

[
|0| < 2.262
]

We **accept the null hypothesis**.

---

### 6. Verification Using Python

Example Python code:

```python
import scipy.stats as stats

ages = [18,19,20,19,18,20,19,19,18,20]

# t-test comparing sample mean with population mean = 19
t_stat, p_value = stats.ttest_1samp(ages, 19)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject H0")
else:
    print("Accept H0")
```

**Expected Result**

* t-statistic ≈ **0**
* p-value ≈ **1**

Since:

[
p\text{-value} > 0.05
]

We **fail to reject (accept) (H_0)**.

Conclusion:
At 95% confidence level, the null hypothesis is accepted, meaning there is no significant difference from 19 years. Therefore, the average age of TE IT students can be considered 19 years---
